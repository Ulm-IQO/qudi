```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

N = 13 # number of data points
t = np.linspace(0, 180, 13)
data = [0.278, 0.066, 0.018, 0.190, 0.404, 0.450, 0.287, 0.081, 0.027, 0.174, 0.442, 0.497, 0.317]

guess_freq = 1/75
guess_amplitude = 3*np.std(data)/(2**0.5)
guess_phase = 180
guess_offset = np.mean(data)

p0=[guess_freq, guess_amplitude,
    guess_phase, guess_offset]

# create the function we want to fit
def my_sinsq(x, freq, amplitude, phase, offset):
    return (np.sin(x * freq + phase) * amplitude + offset)**2

# now do the fit
fit = curve_fit(my_sinsq, t, data, p0=p0)

t2 = np.linspace(0, 180, 100)

# we'll use this to plot our first estimate. This might already be good enough for you
data_first_guess = my_sinsq(t2, *p0)

# recreate the fitted curve using the optimized parameters
data_fit = my_sinsq(t2, *fit[0])


plt.plot(t2,data_fit,'g-', label='after fitting')
plt.plot(t2,data_first_guess,'b-', label='first guess')
plt.plot(t, data, 'bo')

plt.legend()
plt.show()
```
